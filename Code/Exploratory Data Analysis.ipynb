```python
# Importing the libraries 
import pandas as pd
from IPython.display import display
import os
import numpy as np
import time
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Setting up the path for reading file
src = r'C:\Users\Saad.LAKES\Desktop\Pavement-Temperature-Prediction\Data'
filename = 'Pave_data_1.xlsx'

# Setting up the path for the final file
dest = r'C:\Users\Saad.LAKES\Desktop\Pavement-Temperature-Prediction\Data'
destfile = 'Pave_data_cleaned.xlsx'
```

```python
# Reading the file 
df = pd.read_excel(os.path.join(src,filename))

# Displaying the basic information of the data
display(df.head())
display(df.describe())
```

```python
# Determining the number of missing values
df.isnull().sum()
```

```python
# Since the result shows that the missing values are only in the Temp column, they 
# are supposed to be resolved. For that first the parameter is plotted and then interpolated
df['Temp'].plot()
plt.show()

# The missing values are being filled up
df['Temp'].interpolate(method='linear',limit_direction='both',inplace=True)
```

```python
# Since the describe shows that the minimum value of Solar is -999 which is impossible so that is 
# also treated as a missing value and so it is being replaced by the mean since there is only one value
# associated with such a number

# Counting the values associated with -999
df.loc[df['Solar']==df['Solar'].min()]

# Filling up the value with mean
df['Solar']= np.where(df['Solar']==df['Solar'].min(),df['Solar'].mean(),df['Solar'])
```

```python
# Plotting the relationships between all the variables 
sns.pairplot(df)
```

The pairplot illustrates multiple things:
- The temperature and pavement temperature follow a linear relationship 
- The solar radiation has no relationship with the pavement
- The day and hour do have a relationship but it is of small effect
- The day and hour follows almost the same trajectory 
- The pavement itself follows a normal distribution

```python
# Plotting the correlation matrix for all the parameters
plt.figure(figsize=(14, 10))
heatmap = sns.heatmap(df.corr(), vmin=-1, vmax=1, annot=True,cmap="YlGnBu")
heatmap.set_title('Correlation Matrix', fontdict={'fontsize':12}, pad=12);
```

The correlation matrix further supports the claims in the pairplot. 
- It is found that the temperature and pavement is highly correlated 
- The month and year do correlate with pavement temperature but with small effect (|correlation| < 0.3)
- The day does not correlate much with the pavement (correlation $\approx$ 0.0)
- The similar is the case of solar radiation
